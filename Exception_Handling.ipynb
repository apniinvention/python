#Python Exception Handling



####Error in Python can be of two types i.e. 
**Syntax errors** and **Exceptions**.   
* **Errors** are the problems in a program due to which the program will stop the execution. 
* **Exceptions** are raised when some internal events occur which changes the normal flow of the program. 

###How try() works? 
 

First, the try clause is executed i.e. the code between try and except clause.
If there is no exception, then only the try clause will run, except the clause is finished.
If any exception occurs, the try clause will be skipped and except clause will run.
If any exception occurs, but the except clause within the code doesnâ€™t handle it, it is passed on to the outer try statements. If the exception is left unhandled, then the execution stops.
A try statement can have more than one except clause

1. Error in Python

```python
#Example of Error in Python
a = 4
if a % 2 == 0 
  print("Even Number")
```

2. Exception  
 
Exceptions are raised when the program is syntactically correct, but the code resulted in an error. This error does not stop the execution of the program, however, it changes the normal flow of the program.

```python
# Example of Exception : ZeroDivisionError
a = 100

print(a/0)
```

In the above example raised the ZeroDivisionError as we are trying to divide a number by 0.

Let's see how we can handle the exception gracefully  

By handling the exception means either we can gracefully termincate the execution by raising the error with proper message or we can write a second level execution as well (not much recomended)



```python
# Example of a simple exception
# type : Index out of bound

a = [1,2,3]

try :
  print(a[4])
except :
  print(" Error occured because of your logic")
```

In the above example, within the try block we can write our own logic and based on the logic we can raise an exception related to our logic

Now, Lets Catch the specific Exception

```python
#Example of catching specific Exception
a = 100

try:
  print(a/0)
except ZeroDivisionError:
  print("ZeroDivisionError Occurred and Handled")
```

We can also add ELSE cluase after except

```python
# Example of else cluase with try & except
a = 100

try:
  if a != 100:
    print(a/0)
except ZeroDivisionError:
  print("ZeroDivisionError Occurred and Handled")
else:
  print("Exception didnt occured")
```

Finally Keyword 

```python
a = 100

try:
  if a != 100:
    print(a/0)
except ZeroDivisionError:
  print("ZeroDivisionError Occurred and Handled")
else:
  print("Exception didnt occured")
finally:
  print("This will always executed")
```
