# Python Function


* Python Functions is a **block of statements that return the specific task**.  

* The idea is to put some **commonly or repeatedly done tasks together** and make a function so that instead of writing the same code again and again for different inputs, we can do the function calls to **reuse code contained in it over and over again**.

```python
#First Python Function

# 1. Create a python function to add two number

def add(a,b):
  result = a + b
  return result

#function call
output = add(2,3) # 2,3 are the parameter of the function

#Function return the value to its caller, here add(2,3) is the caller place, 
#once result is computed, in place add(2,3) consider the result of the function i.e 5 & finally "5" is assigned to its variable i.e "output"


print(output)
```

```python
#Second Python Function

# 2. Create a python function similar to java or C/C++ where we mention the data type of variable & its return type

def add(a: int, b: int) ->int:
  result = a + b
  return result

output = add(2,3)
print(output)

output = add(2.1,3.5)
print(output)
```

###Arguments of a Python Function  
Arguments are the values passed inside the parenthesis of the function. A function can have any number of arguments separated by a comma.

```python

# a,b : is the argument of the python function
# Point to remember : Number argument in function defination & from where you are calling should match

def add(a,b):
  result = a + b
  return result


output = add(2,3)

print(output)
```

## Types of Arguments
* Default arguments : A default argument is a parameter that assumes a default value if a value is not provided in the function call for that argument.

* Keyword arguments : The idea is to allow the caller to specify the argument name with values so that caller does not need to remember the order of parameters.  

* Variable-length arguments :  In Python, we can pass a variable number of arguments to a function using special symbols. 
  * There are two special symbols:
    * *args (Non-Keyword Arguments)
    * **kwargs (Keyword Arguments)



 


```python
# Python Prgoram to demonstrate Default Arguments

def add(a,b,c=0):
  result = a+b+c
  return result

#Calling the function with only two argument, but in the defination we have assigned a default value, so it handle automatically
output = add(2,3)
print("The output of two parameter --> ",output)

#Bacause we are providing the third parameter, it will not take the default value, it will use the caller value of c i.e 4 
output = add(2,3,4)
print("The output of three parameter --> ",output)
```

```python
# Python Program to Demonstrate "Keyword" argument


def add(first_value,second_value):
  result = first_value + second_value
  return result

# As we are providing the value in the function with the help of "=", we have to make sure that the argument name is same in both the places : defination & caller both
output = add(first_value = 2, second_value=3)
print(output)

# We generaally use this concept when we have two many argument or for readibilty purpose as well
```

```python
#Variable Length Argument

# Type 1: *args : Non-Keyword argument

def print_seq(*argv):
  for each in argv:
    print(each)

print_seq(1,2,3,4)

#We can use any name for this argument
print()
print("Using Hello as argument name")
def print_seq(*hello):
  for each in hello:
    print(each)

print_seq(1,2,3,4)
```

```python
#Variable Length Argument

# Type 2: **kwargs : Keyword Arguments

def print_seq(**kwargs):
  for key, value in kwargs.items():
    print(key,value)

print_seq(first_num=1, second_num=2, third_num=3)
```

##Anonymous functions in Python Function
An anonymous function means that a function is without a name 
*  def keyword is used to define the normal functions
* lambda keyword is used to create anonymous functions



```python
# Example of An anonymous function 

print("Using Lambda Function ")
square = lambda x : x*x
print(square(2))

print("using normal python function")
def square(x) : return x*x 
print(square(2))

#The above function is equivalent to

# def square(x):
#   return x*x
```
